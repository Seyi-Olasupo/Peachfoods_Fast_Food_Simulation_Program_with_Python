## Peachfoods, Aberdeen.

Peachfoods is a new business in fast-food service in Aberdeen. The aim of this program is to design the business sales transaction simulator.

### Peachfoods Fast-Food Simulation Program: Option and Uses

1. **Admin login**: This option simulate how the user validates their credentials. When this option is selected, the user will be prompted to input their username and password. The username should be *admin* and the password should be any number smaller than *10* (including float and negative numbers). User are permitted to four entry, if all entry is wrong, then the program stops and the user program will need to start all over again.

  * Note: Once an admin has logged in, this option will be disabled (i.e. the user cannot go back and select this option).
  

2. **Add products**: This option can only be accessed once option 1 has been completed, otherwise the message *login first!* will be displayed and the program will go back to the main menu. In this option, the admin will be shown the list of foods and categories in the fast-food menu. Then, admin (user) can input a menu number  to add *one more product* for that certain menu slot. Afterwards, the program will show the new foods and go back to the main menu.
  
  * Note: The each space in the fast-food menu can hold a maximum of *eight* products (of the same food, only). Also, products cannot be added to empty menu slots (NAN), as the will throw a warning note.
  
  
3. **Buy food**: This option can be accessed by "anyone" i.e both admin and the customer, therefore no verfication is run here. If this option is selected, the user will be shown the list of foods, amounts and prices. Then, the user will be requested to select one food item based on the `menu_number`. Once the food is selected, the program display the price of the selected product. Then, the user will be prompted to pay. To simulate this payment, the program ask the user for an input and write any positive number (if the user inputs something invalid, this shows "try again"). Then, the program check the amount input against the price of the product bought. If the input number is larger than the price, then the program response *your change is...* and the subtraction of the payment minus the price of the food. If the amount is equal to the price, the program returns *thanks for paying*. Else, the program output *you need to pay more* and allow the user to write another amount. After the "purchase", the program output to the user(admin) the number of products left for the particular food. 
    
  * Note: If there is zero products left for the food selected, the program prompt this to the user/customer in advance before letting them buy, and ask them to select another food.
  

4. **Exit**

```python
loop = [False, False]
def mainpage():
    print('Welcome to the RGyUm fast-food menu. Select an option:')
    print('1. Log in (admin only)')
    print('2. Add one more product of a certain food (admin only)')
    print('3. Buy products')
    print('4. Exit')
    try:
    
        option=int(input("Enter your option: "))

        if option == 1:
            log_in()
        elif option == 2:
            add_product()
        elif option == 3:
            buy_food()
        elif option == 4:
            exit()
        else:
            print("Invalid option. Select 1-4")
            mainpage()
    except:
        print("\nWrong entry")
        mainpage()

## ADMIN LOGIN EXPLANATION
## The below code program respond to admin login. This code made use of counter, while loop, input statement, float, def.
## If login condition are met it proceed to add product, if not, it display wrong entries.
## If this wrong entries continue for four times, it goes back to mainpage.

def log_in():
    print("Hello")
    try:
        counter = 0
        while counter < 4:
            username = input("Enter your username: ")
            password = input("Enter your password: ")
            if username == "admin" and float(password)<10:
                print("Hi Admin, Welcome.")
                loop[0] = True
                add_product()
                break
            else:
                print(" ")
                print("Wrong entries, Please re_try.")
                print(" ")
                counter = counter+1
            if counter ==4:
                mainpage()
    except:
        print("\nTry to type again")
        mainpage

## ADD_PRODUCT EXPLANATION
## After successful login, the admin can add a product if the product is found in the dataframe.
## If not, the program dispalys "Incorrect menu".
## If the product is NAN in the dataframe, it displays menu not available.
## if product in the df is already 8 in total in store, its write "Product fully stocked".
## Finally if all this conditions and others are met, it successfully add the product to the store.

def add_product():
    if loop[0] == True:
        print(fooditem[['food', 'category']])
        product_number = input("Add product menu number: ").upper()
        if product_number not in fooditem.index:
            print("Incorrect menu number")
            add_product()
        else:
            if pd.isnull(fooditem.loc[product_number, "food"]):
                print("This menu is not available")
                add_product()        
            elif fooditem.loc[product_number, "amount"] == 8:
                print("Products fully stocked!!!")
                add_product()
            else:
                fooditem.loc[product_number, "amount"] = fooditem.loc[product_number, "amount"] +1
                print (fooditem)
                loop[0] = False
                mainpage()
    else:
        print("Login first")
        mainpage()

## BUY_FOOD EXPLANATION
## Here, the program is to buy food.
## If order is 0 in store, the code displays "we are out of stock".
## If order is NAN, the code displays "Empty slot".
## If all the program conditons are met, it successfully allow the user to buy the food they ordered and give them change.
## If the payment is over the price, it display your change is and give the balance change.
## If the payment is below the price, the code will respond with "Invalid amount" and request for payment entry again for balance remaining.
## If all condition are satisfied, the program respond with "Thanks for paying" and give total number of product left in store.
## Then, the order can be successfullly ended if no order request again.

def buy_food():        
    print('welcome')
    print(fooditem[['food', 'amount', 'price']])
    print('\nHere is our Menu List, what will you like to buy?\n')
    order = input('Please enter your menu_number: ').upper()
    if order in fooditem.index:
        if fooditem.loc[order, "amount"] == 0:
            print('We are out of stock, please select another food')
            buy_food()
        elif pd.isnull(fooditem.loc[order, "amount"]):
            print('You selected an empty slot')
            buy_food()
        else:
            loop = True
            while loop:
                print('The price is ' + str(fooditem.loc[order, "price"]))
                try:
                    pay = float(input('\nEnter payment: '))
                    if pay <= 0:
                        print("Invalid amount, please try again.")
                    elif pay == fooditem.loc[order, "price"]:
                        print('Thanks for paying')
                        loop = False
                    elif pay > fooditem.loc[order, "price"]:
                        print('Your change is ' + str(pay - fooditem.loc[order, "price"]))
                        loop = False
                    else:
                        complete_payment = pay
                        while True:
                            try:
                                print('You need to pay ' + str(fooditem.loc[order, "price"] - pay))
                                balance = float(input('Complete payment: '))
                                if balance <= 0:
                                    print("Invalid amount, please try again.")
                                    continue
                                complete_payment += balance
                                if complete_payment > fooditem.loc[order, "price"]:
                                    print('Thanks for paying')
                                    loop = False
                                    break
                                elif complete_payment < fooditem.loc[order, "price"]:
                                    print('You need to pay ' + str(fooditem.loc[order, "price"] - complete_payment))
                                else:
                                    print('Your change is ' + str(complete_payment - fooditem.loc[order, "price"]))
                                    loop = False
                                    break
                            except ValueError:
                                print("Invalid input, please try again.")
                except ValueError:
                    print("Invalid input, please try again.")
            
        fooditem.loc[order, "amount"] = fooditem.loc[order, "amount"] - 1
        print('There is only ' + str(fooditem.loc[order, "amount"]) + ' ' + str(fooditem.loc[order, "food"]))
        print(fooditem)
        mainpage()
        
    else:
        print("Invalid menu number")
        buy_food()

def exit():
    runtime.unassign()
    
mainpage()
exit()
```
